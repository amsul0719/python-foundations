```python
# Load CellProfiler Excel export - one row per well, contains imaging metrics 
df_cellprofiler = pd.read_excel("cellprofiler_export.xlsx", sheet_name="Well_Data")

df_cellprofiler
```

## 3. Initial QC — Check for Missing Values

```python
# Count missing values per column - CellProfiler data
df_cellprofiler.isnull().sum()
```

```python
# Filter to failed wells - rows where TotalCells_Count is missing
failed_mask=df_cellprofiler["TotalCells_Count"].isnull()
df_cellprofiler[failed_mask]
```

## 4. Remove Failed Wells — Create Clean Dataset

```python
# Exclude failed wells — keep only rows where TotalCells_Count is not missing
# ~ flips the mask: True becomes False and False becomes True
# Original df_cellprofiler is preserved; df_clean is used for downstream analysis
clean_mask = ~ failed_mask
df_clean = df_cellprofiler[clean_mask]

df_clean = df_cellprofiler[clean_mask].copy()
df_clean
```

```python
df_clean["MBP_ratio"]=df_clean["MBP_Positive_Count"] / df_clean["TotalCells_Count"]
```

```python
df_clean
```

```python
DMSO_mask = df_clean["Compound"] == "DMSO" 

df_DMSO = df_clean[DMSO_mask]
df_DMSO
```

```python
DMSO_avg = df_DMSO["MBP_ratio"].mean()
DMSO_avg
```

```python
df_clean["MBP_ratio_normalized"] = df_clean["MBP_ratio"] / DMSO_avg
df_clean
```

```python
potent_compound = df_clean["MBP_ratio_normalized"] > 10

df_clean[potent_compound]
```
