## Core Operators

| Symbol | Meaning | Description |
|--------|----------|------------|
| `&` | AND | True only if both conditions are True (row-wise) |
| `|` | OR | True if at least one condition is True (row-wise) |
| `~` | NOT | Flips True ↔ False in a Boolean mask |

---

## Rule Reminder

When using `&` or `|`, always wrap each condition in parentheses:

Correct:
```python
(df["effect"] > 20) & (df["toxicity"] < 50)

````python
import pandas as pd
````

````python
#--- Parameters/ example data (defined first) ---
compounds = ["A", "B", "C", "D"]
effect_values = [ 10, 35, 50, 15]
toxicity_values = [20, 60, 30, 10]

#--- Build Data Frames ---
df = pd. DataFrame({
    "compound": compounds, 
    "effect": effect_values, 
    "toxicity": toxicity_values 
})

df
````

````python
#--- Thresholds (defined first) ---
effect_threshold = 20
toxicity_threshold = 50
````

````python
#--- Single-condition masks ---
active_mask = df["effect"] > effect_threshold
safe_mask = df["toxicity"] < toxicity_threshold

active_mask, safe_mask
````

````python
#--- Combined Mask ---
active_and_safe_mask = active_mask & safe_mask
active_and_safe_mask 
````

````python
df["active"] = active_mask
df["safe"] = safe_mask
df["active_and_safe"] = active_and_safe_mask

df
````

````python
df["decision"] = df["active_and_safe"].map({True: "Advance", False: "Drop"})
df
````
